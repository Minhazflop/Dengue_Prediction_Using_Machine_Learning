```python
# Convert numeric columns properly
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])  # Convert only valid numeric columns
    except ValueError:
        pass  # Skip non-numeric columns

# Function to detect outliers using IQR method and return detailed info
def detect_outliers(data):
    Q1 = data.quantile(0.25)  # First quartile
    Q3 = data.quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': len(outliers),
        'Outliers': outliers.tolist()
    }

# Apply function to all numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
outlier_results = {col: detect_outliers(df[col]) for col in numeric_cols}

# Display detailed results
print("Outlier Detection Summary (IQR Method):")
print("-" * 50)
for col, result in outlier_results.items():
    print(f"\nFeature: {col}")
    print(f"Q1: {result['Q1']:.2f}, Q3: {result['Q3']:.2f}, IQR: {result['IQR']:.2f}")
    print(f"Lower Bound: {result['Lower Bound']:.2f}, Upper Bound: {result['Upper Bound']:.2f}")
    print(f"Number of Outliers: {result['Outlier Count']}")
    if result['Outlier Count'] > 0:
        print(f"Example Outliers (up to 5): {result['Outliers'][:5]}")

# Generate colorful boxplots for all numeric features
plt.figure(figsize=(15, 10))
colors = plt.cm.Set3(np.linspace(0, 1, len(numeric_cols)))  # Use Set3 colormap for variety

for i, col in enumerate(numeric_cols, 1):
    plt.subplot((len(numeric_cols) + 2) // 3, 3, i)  # Dynamic grid layout
    
    # Create the boxplot with custom styling
    bp = df.boxplot(column=col, patch_artist=True, 
                    return_type='dict',  # Return boxplot components for customization
                    vert=True)
    
    # Customize box colors
    for patch, color in zip(bp['boxes'], [colors[i-1]]):
        patch.set_facecolor(color)  # Fill box with color
        patch.set_alpha(0.7)  # Slight transparency for better overlap visibility
    
    # Customize whiskers (purple)
    for whisker in bp['whiskers']:
        whisker.set(color='purple', linewidth=1.5)
    
    # Customize caps (orange)
    for cap in bp['caps']:
        cap.set(color='orange', linewidth=1.5)
    
    # Customize medians (red)
    for median in bp['medians']:
        median.set(color='red', linewidth=2)
    
    # Customize outliers (green dots)
    for flier in bp['fliers']:
        flier.set(marker='o', color='green', alpha=0.5, markersize=5)
    
    plt.title(f'Boxplot of {col}', fontsize=10, color='darkblue')
    plt.grid(True, linestyle='--', alpha=0.3)  # Add a light grid for reference

plt.tight_layout()
plt.show()
```
